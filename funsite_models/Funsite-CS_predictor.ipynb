###  FunSite and generic predictions from 5-fold cross-validation of the CS training dataset

```python
def get_pr_curve_model_pred_dataset(X, y, groups):
    
    y_real=[]
    y_proba=[]
    i = 0
    precision_array = []
    threshold_array=[]
    recall_array = np.linspace(0, 1, 100)
    auc_array=[]
    
    for train, test in group_kfold.split(X, y, groups):
        
        X_train = X.iloc[train]
        y_train = y.iloc[train]
        X_test = X.iloc[test]
        y_test = y.iloc[test]

        classifier = XGBClassifier(

            n_estimators = 1000,
            learning_rate = 0.01,
            subsample = 0.8,
            max_depth=7,
            colsample_bytree = 1, 
            objective = 'binary:logistic',
            scale_pos_weight = 6,
            n_jobs= -1,
            random_state=0,

        )
        
        classifier.fit(X_train, y_train)
        pred_proba = classifier.predict_proba(X_test)
    
        # Compute PR curve and area the curve
        precision, recall, thresholds = precision_recall_curve(y_test, pred_proba[:, 1])
        precision, recall, thresholds = precision[::-1], recall[::-1], thresholds[::-1]  # reverse order of results

        thresholds = np.insert(thresholds, 0, 1.0)
        
        precision_array = interp(recall_array, recall, precision)
        threshold_array = interp(recall_array, recall, thresholds)
        pr_auc = auc(recall_array, precision_array)
 
        y_real.append(y_test)
        y_proba.append(pred_proba[:,1])
        auc_array.append(auc(recall, precision))

        i += 1
        
    y_real = np.concatenate(y_real)
    y_proba = np.concatenate(y_proba)

    precision_overall, recall_overall, _ = precision_recall_curve(y_real, y_proba)
    pr_auc_overall = auc(recall_overall, precision_overall)
    
    mean_precision = np.mean(precision_array)
    std_precision = np.std(precision_array)

    return (precision_overall, recall_overall, pr_auc_overall)
```

```python
precision_ff, recall_ff, pr_auc_ff = get_pr_curve_model_pred_dataset(X_ff, y, groups)
precision_generic, recall_generic, pr_auc_generic = get_pr_curve_model_pred_dataset(X_generic, y, groups)
```

```python
plt.rcdefaults()
plt.rcParams['figure.figsize'] = (5.0, 5.0)
plt.rcParams['axes.facecolor']='white'

label_ff = 'FunSite-CS (PR AUC=%.3f)'%(pr_auc_ff)
label_generic = 'Generic-CS (PR AUC=%.3f)'%(pr_auc_generic)

plt.plot(recall_ff, precision_ff, lw=2,color='red', label=label_ff)
plt.plot(recall_generic, precision_generic, lw=2,color='grey', label=label_generic)

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='lower left', fontsize='medium')

# plt.savefig('FunSite-CS_vs_generic.png', dpi=300, bbox_inches='tight')
plt.show()
```
